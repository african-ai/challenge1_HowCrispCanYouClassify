# Classification

## Decision Trees

```python
Train_digitsdf = pd.read_csv("digits_train.csv").values
Test_digitsdf = pd.read_csv("digits_test.csv").values
```

```python
#Training Data 
clf = DecisionTreeClassifier()
xtrain = Train_digitsdf[0:3000:, 1:65]
ytrain = Train_digitsdf[0:3000:, -1]
```

```python
clf.fit(xtrain, ytrain)
```

```python
#Validation Data
xtest = Train_digitsdf[3000:, 1:65]
Actual_Labels = Train_digitsdf[3000:, -1]
```

```python
y_predict = clf.predict(xtest)
len(Actual_Labels)
```

```python
count = 0
for i in range(0, 1000):
    count += 1 if y_predict[i]==Actual_Labels[i] else 0
print("The prediction accuracy for Decision tree is=", (count/1000)*100,"%")
```

```python
print ("The prediction accuracy for Decision tree is = ", metrics.accuracy_score(y_predict, Actual_Labels))
```

## K Nearest Neighbors

```python
model = KNeighborsClassifier(n_neighbors=3) # this examines 3 neighbors for putting the data into class
model.fit(xtrain, ytrain)
y_predict = model.predict(xtest)
print('The prediction accuracy for KNN is: ', metrics.accuracy_score(y_predict, Actual_Labels))
```

## Logistics Regression

```python
model = LogisticRegression()
model.fit(xtrain, ytrain)
y_predict = model.predict(xtest)
print('The prediction accuracy for LG is: ', metrics.accuracy_score(y_predict, Actual_Labels))
```
