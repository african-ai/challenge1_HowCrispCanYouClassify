# Classification of different types of iris flower


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Load the datasets

```python
iris_train = pd.read_csv('data/iris_train.csv', index_col=0)
iris_test = pd.read_csv('data/iris_test.csv', index_col=0)
```

```python
iris_train.head()
```

```python
iris_test.head()
```

### Visualize the dataset relationships

```python
sns.pairplot(iris_train, hue='labels')
```

### Setting feature matrix and target array

```python
X_train = iris_train.iloc[:, :-1].values
y_train = iris_train.iloc[:, -1].values
```

### Model selection

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifier_model = GaussianNB()
```

### Model fitting

```python
classifier_model.fit(X_train, y_train)
```

### Prediciting the results

```python
y_pred = classifier_model.predict(iris_test.values)
```

```python
y_pred
```

```python
y_pred_df = pd.DataFrame(y_pred)
```

```python
y_pred_df.to_csv('submissions/iris_test_labelled.csv')
```
