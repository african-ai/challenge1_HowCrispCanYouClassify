```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("data/digits_train.csv", index_col=0)
```

```python
train_df.head()
```

```python
test_df = pd.read_csv("data/digits_test.csv", index_col=0)
```

```python
test_df.describe()
```

```python
test_data = np.array(test_df)
```

```python
# drop labels to get feature
y = train_df['labels']
X = train_df.drop('labels', axis=1)
```

```python
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2)
```

```python
classifier = RandomForestClassifier(n_estimators=10, criterion = 'entropy')
classifier.fit(X_train, y_train)
print ('Accuracy of Random forest classifier on training set: {:.2f}'.format(classifier.score(X_train, y_train)))
print('Accuracy of Random forest classifier on test set: {:.2f}'.format(classifier.score(X_test, y_test)))
```

```python
# predict test data
predictions = classifier.predict(test_df)
```

```python
predictions
```

```python
test_pred = pd.DataFrame()
test_pred['labels'] = predictions
test_pred.to_csv('submissions/submission_digit.csv', index=True)
```
